```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
print(country_name)
```

```python
country_val=final_df.Country.value_counts().values
print(country_val)
```

```python
#Pie Chart -Top 3 countries that use Zomato
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation:Zomato's maximum transactions are from India,followed by USA then UK

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
ratings
```

Observation: When ratings are between
4.5 to 4.9 = Excellent
4.0 to 4.4 = Very Good
3.5 to 3.9 = Good
3.0 to 3.4 = Average
2.5 to 2.9 = Average
2.0 to 2.4 = Poor


```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Counts",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Counts",hue="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observations:
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
#Finding country names that have given 0 ratings
final_df.groupby(['Aggregate rating','Country']).size().reset_index().head()
```

Observations: Maximum number of 0 ratings are from Indian Customers

```python
#finding out which currency is used by which country
final_df.groupby(['Country','Currency']).size().reset_index()
```

```python
#finding which countries have online deliveries option
final_df[['Has Online delivery','Country']].groupby(['Has Online delivery','Country']).size().reset_index()
```

Observations:Online Deliveries are available in India and UAE
